```python
import os
import glob

folder_path = "/config/workspace/Weather_pattern/MORA_report"
# Get a list of all files in the folder
files = glob.glob(os.path.join(folder_path, '*.xlsx'))
# Sort the files by modification time (latest first)
files.sort(key=os.path.getmtime, reverse=True)

if files:
    latest_file = files[0]
    print("The latest file in the folder is:", latest_file)
else:
    print("The folder is empty or doesn't exist.")


```

```python
Node_MORA = pd.read_excel(latest_file,sheet_name="Resource Details",header=1)
print(Node_MORA.shape)
Node_MORA=Node_MORA[~Node_MORA['UNIT CODE'].isna()]
print(Node_MORA.shape)

Node_MORA =  Node_MORA[~Node_MORA['COUNTY'].isna()]
print(Node_MORA.shape)
Node_MORA =  Node_MORA[ Node_MORA.COUNTY.str.isalpha()]
print(Node_MORA.shape)
Node_MORA['COUNTY_texas']= Node_MORA['COUNTY']+ ',texas'
```
